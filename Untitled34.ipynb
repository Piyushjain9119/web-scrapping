```python
 amazon_page = client.read()
```

```python
amazon_html = bf(amazon_page,'html.parser')
```

```python
bigbox = amazon_html.findAll("div",{"class":"sg-col-inner"})
```

```python
len(bigbox)
```

```python
list1 = []
for i in bigbox:
    print(i)
    try:
        urls ="https://www.amazon.in/"+i.div.div.div.div.div.div.span.a['href']
        list1.append(urls)
    except Exception as e :
        e
```
